```python
data["data_invio_candidatura"] = pd.to_datetime(data["data_invio_candidatura"], errors='coerce')
data["data_finanziamento"] = pd.to_datetime(data["data_finanziamento"], errors='coerce')
```

```python
finanziamenti_per_regione = data.groupby("regione")["importo_finanziamento"].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(x=finanziamenti_per_regione.values, y=finanziamenti_per_regione.index, palette="viridis")
plt.xlabel("Importo Totale Finanziato (€)")
plt.ylabel("Regione")
plt.title("Distribuzione dei Finanziamenti per Regione")
plt.show()
```

```python
# --- Numero di Comuni Finanziati per Regione ---
comuni_finanziati_per_regione = data.groupby("regione")["comune"].nunique().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=comuni_finanziati_per_regione.values, y=comuni_finanziati_per_regione.index, palette="coolwarm")
plt.xlabel("Numero di Comuni Finanziati")
plt.ylabel("Regione")
plt.title("Numero di Comuni Finanziati per Regione")
plt.show()
```

```python
import geopandas as gpd
# --- Mappa di Calore della Distribuzione dei Finanziamenti ---
italy_map = gpd.read_file("https://raw.githubusercontent.com/openpolis/geojson-italy/master/geojson/limits_IT_regions.geojson")

# Unire i dati con la mappa
italy_map = italy_map.merge(finanziamenti_per_regione, left_on="reg_name", right_index=True, how="left")

# Creare la mappa di calore
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
italy_map.plot(column="importo_finanziamento", cmap="Reds", linewidth=0.8, edgecolor='black', legend=True, ax=ax)
plt.title("Distribuzione dei Finanziamenti per Regione (Mappa di Calore)")
plt.show()
```

```python
# --- Analisi Temporale delle Candidature ---
candidature_per_mese = data.resample('M', on="data_invio_candidatura").size()

plt.figure(figsize=(12, 6))
sns.lineplot(x=candidature_per_mese.index, y=candidature_per_mese.values, marker="o", color='orange')
plt.xlabel("Mese")
plt.ylabel("Numero di Candidature")
plt.title("Andamento delle Candidature nel Tempo")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# --- Analisi delle Tempistiche Medie tra Invio e Approvazione ---
# Ensure both datetime columns are timezone-naive
data["data_invio_candidatura"] = data["data_invio_candidatura"].dt.tz_localize(None)
data["data_finanziamento"] = data["data_finanziamento"].dt.tz_localize(None)

data["tempo_attesa"] = (data["data_finanziamento"] - data["data_invio_candidatura"]).dt.days

# Calcolare la media dei giorni di attesa per regione
tempistiche_medie = data.groupby("regione")["tempo_attesa"].mean().sort_values()

plt.figure(figsize=(12, 6))
sns.barplot(x=tempistiche_medie.values, y=tempistiche_medie.index, palette="Blues")
plt.xlabel("Giorni Medi di Attesa")
plt.ylabel("Regione")
plt.title("Tempistiche Medie tra Invio e Approvazione del Finanziamento")
plt.show()
```

```python
# --- Identificazione dei Periodi di Picco per l'Invio delle Candidature ---
candidature_per_giorno = data.resample('D', on="data_invio_candidatura").size()

plt.figure(figsize=(12, 6))
sns.lineplot(x=candidature_per_giorno.index, y=candidature_per_giorno.values, color='red')
plt.xlabel("Data")
plt.ylabel("Numero di Candidature")
plt.title("Periodi di Picco per l'Invio delle Candidature")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# --- Identificazione dei Comuni con i Finanziamenti Più Alti ---
top_comuni = data.groupby("comune")["importo_finanziamento"].sum().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_comuni.values, y=top_comuni.index, palette="magma")
plt.xlabel("Importo Finanziato (€)")
plt.ylabel("Comune")
plt.title("Top 10 Comuni con i Finanziamenti più Alti")
plt.show()
```
